```python

li = [1,2,3,4,5,6,7]
#normal way
for i in range(len(li)-1,-1,-1):
    print(li[i],end =" " )
print()   
  #idiomatic way  
for i in reversed(li):
    print(i,end= " ")
```

```python
li = [1,2,3,4,5,6,7]
for i,a in enumerate(li):
    print(i,"-->",a)
```

```python
li = [1,2,3,4,5,6]
a = [1,2,4,5,6,7,9,3,0,9,12]
n = min(len(li),len(a))
for i in range(n):
    print(li[i],' ',a[i])
print()

for i,a in zip(li,a):
    print(i,' ',a)
```

# Lambda Functions
- Anonymous function means that a function is without a name

```python
a = lambda x : x * x
print(a(10))
```

```python
## Lambda with Filter

def filterList(li):
    a=[]
    for i in li :
        if i % 2 == 0:
            a.append(i)
    return a
li = [1,2,3,4,5,6,7,8]
filterList(li)
```

```python
li = [1,2,4,5,6,7,8]
lam_i = list(filter(lambda x: (x%2==0),li))
print(lam_i)
```

```python
def squareList(li):
    a = []
    for i in li:
        a.append(i*2)
    return a    
li = [1,2,3,4,5,6,7,8,9]
squareList(li)
```

```python
li = [1,2,3,4,5,6,7,8,9]
map_list = list(map(lambda x : x*2,li))
print(map_list)
```

```python
def sumList(li):
    s = 0
    for i in li:
        s += i
    return s
li = [1,2,3,4,5,6,7,8,9]
sumList(li)
```

```python
from functools import reduce
li = [1,2,3,4,5,6,7,8,9]
s = reduce((lambda x,y:x+y),li)
print(s)
```
